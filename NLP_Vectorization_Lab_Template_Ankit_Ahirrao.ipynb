## First Part - Toy Dataset

### Toy Dataset
The following toy dataset is used for demonstrating text preprocessing and vectorization techniques.



```python
corpus = [
    "machine learning is fun",
    "machine learning is powerful",
    "deep learning is fun",
    "deep learning powers modern ai"
]
```

## 3. Preprocessing

The following preprocessing steps are applied to the text data:
- Convert all text to lowercase
- Remove punctuation
- Tokenize the text into words
- Remove stopwords (optional)



```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def preprocess_text(text):
    text = text.lower() #Lowercasing
    tokens = re.findall(r'\b[a-z]+\b', text) # Regex-based tokenisation
    words = [word for word in tokens if word not in ENGLISH_STOP_WORDS] # stopword removal
    return ' '.join(words)

preprocessed_corpus = [preprocess_text(doc) for doc in corpus]
preprocessed_corpus
```

## 4. One-Hot Encoding

One-hot encoding represents each word as a binary vector indicating its presence or absence in a document.


## 4. One-Hot Encoding


```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary=True)
X_onehot = vectorizer.fit_transform(preprocessed_corpus)

print("Vocubulary:", vectorizer.get_feature_names_out())
print("One-Hot Encoded Matrix:")
print(X_onehot.toarray())
```

## 5. Bag of Words (BoW)

Bag of Words represents text by counting the frequency of each word in a document.


```python
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(preprocessed_corpus)

print("Vocubulary:", vectorizer.get_feature_names_out())
print("Bag of Words Matrix:")
print(X_bow.toarray())
```

## 6. TF-IDF

TF-IDF assigns higher weight to important words and reduces the weight of common words.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(preprocessed_corpus)

print("Vocubulary:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:")
print(X_tfidf.toarray())
```

## 7. Similarity Analysis

Cosine similarity is used to measure similarity between documents.


```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(X_tfidf)
similarity_matrix
```

```python
## Finding most similar document pair

import numpy as np
np.fill_diagonal(similarity_matrix, 0)
max_similarity = np.unravel_index(similarity_matrix.argmax(), similarity_matrix.shape)
print("Most similar documents indexes:", max_similarity)

## Mapping index to actual document text
print("Document 1:", preprocessed_corpus[max_similarity[0]])
print("Document 2:", preprocessed_corpus[max_similarity[1]])
```

## 8. Comparative Study
| Feature | One-Hot Encoding | Bag of Words (BoW) | TF-IDF |
|-------|------------------|-------------------|--------|
| Representation | Binary (0 or 1) | Frequency-based | Weighted (TF × IDF) |
| Word Order | Ignored | Ignored | Ignored |
| Vocabulary Size | Large | Large | Large |
| Sparsity | Very High | Very High | Very High |
| Captures Frequency | No | Yes | Yes |
| Captures Importance | No | No | Yes |
| Effect of Common Words | Same as rare words | Dominates counts | Down-weighted |
| Use Case | Simple presence check | Basic text analysis | Information retrieval, similarity |




## 9. Observations

1. Text preprocessing such as lowercasing, punctuation removal, tokenization, and stopword removal significantly reduces noise in the dataset and improves the quality of feature extraction.

2. One-Hot Encoding represents words using binary values, indicating only the presence or absence of terms. This results in highly sparse vectors and does not capture word frequency.

3. Bag of Words (BoW) captures the frequency of words in documents, providing more information than One-Hot Encoding. However, it does not consider the importance of words across documents.

4. TF-IDF assigns higher weights to important and less frequent words while reducing the impact of commonly occurring words. This leads to better feature representation.

5. All vectorization techniques produce sparse matrices due to the large vocabulary size compared to the number of words in each document.

6. Cosine similarity effectively measures the similarity between documents by comparing their vector representations.

7. TF-IDF-based similarity analysis produces more meaningful similarity results compared to One-Hot Encoding and BoW, as it emphasizes informative words.

8. Documents containing similar keywords such as “learning” and “fun” show higher similarity scores.


## 10. Conclusion

In this experiment, different text vectorization techniques such as One-Hot Encoding, Bag of Words, and TF-IDF were implemented and analyzed using a toy dataset. Text preprocessing steps helped in cleaning the data and improving feature representation.

It was observed that One-Hot Encoding only captures the presence of words, while Bag of Words considers word frequency but ignores term importance. TF-IDF proved to be more effective as it assigns higher weights to important words and reduces the influence of commonly occurring terms.

Cosine similarity was used to measure document similarity, and TF-IDF-based similarity analysis produced more meaningful and accurate results. Overall, TF-IDF is more suitable for text representation and similarity analysis in practical natural language processing applications.


## Second Part - Real World Dataset (Amazon Product Reviews)

### 1. Dataset Description

The Amazon Product Reviews dataset is used as a real-world textual dataset.
It contains customer-written reviews expressing opinions about products.

A small subset of 300 reviews is selected to demonstrate preprocessing,
text vectorization techniques, and similarity analysis on real-world data.

### 2. Load Dataset

```python
import pandas as pd

df = pd.read_csv("/content/sample_data/Reviews.csv",
                 engine="python",
                 on_bad_lines="skip",
                 quoting=3)

df = df[['Text']].dropna()

df = df.sample(300, random_state=42)

corpus = df['Text'].tolist()

print("Number of reviews loaded:", len(corpus))
```

```python
corpus[:3]
```

### 3. Preprocessing

The following preprocessing steps are applied to the text data:

- Convert all text to lowercase to ensure uniformity.
- Perform regex-based tokenization to extract valid word tokens.
- Remove special characters and punctuation during tokenization.
- Remove common English stopwords to reduce noise and improve feature quality.


```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def preprocess(text):
  text  = text.lower()
  tokens = re.findall(r'\b[a-z]+\b', text) #regex tokenization
  tokens = [w for w in tokens if w not in ENGLISH_STOP_WORDS] #stop word removal
  return " ".join(tokens)

preprocessed_corpus = [preprocess(doc) for doc in corpus]
preprocessed_corpus[:3]
```

###4. One-Hot Encoding
One-hot encoding represents each word as a binary vector indicating the presence or absence of a word in a document.

```python
from sklearn.feature_extraction.text import CountVectorizer

onehot_vectorizer = CountVectorizer(binary=True, max_features=5000)
X_onehot = onehot_vectorizer.fit_transform(preprocessed_corpus)

print("Vocabulary:", onehot_vectorizer.get_feature_names_out())
print("One-Hot Shape:", X_onehot.shape)
```

### 5. Bag of Words (BoW)

Bag of Words represents documents using the frequency of each word.


```python
bow_vectorizer = CountVectorizer(max_features=5000)
X_bow = bow_vectorizer.fit_transform(preprocessed_corpus)

print("BoW Shape:", X_bow.shape)
```

### 6. TF-IDF

TF-IDF assigns higher weights to important words and reduces the impact of commonly occurring words.


```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf_vectorizer.fit_transform(preprocessed_corpus)

print("TF-IDF Shape:", X_tfidf.shape)
```

### 7. Similarity Analysis

Cosine similarity is used to measure similarity between documents.


```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

similarity_matrix = cosine_similarity(X_tfidf)

#Remove self-similarity
np.fill_diagonal(similarity_matrix, 0)

#Find most similar document pair
max_similarity = np.unravel_index(similarity_matrix.argmax(), similarity_matrix.shape)
print("Most similar documents indexes:", max_similarity)
```

```python
i, j = max_similarity
print("Document 1:", corpus[i])
print("Document 2:", corpus[j])
```

### 8. Comparative Study

| Technique | Representation | Observation |
|---------|----------------|-------------|
| One-Hot Encoding | Binary | Very sparse and high dimensional |
| Bag of Words | Frequency-based | Common words dominate |
| TF-IDF | Weighted | Best for similarity analysis |


### 9. Observations

1. Real-world text data contains noise and requires preprocessing.
2. Vocabulary size increases significantly, leading to sparse vectors.
3. One-Hot Encoding is inefficient for large datasets.
4. Bag of Words captures frequency but ignores importance.
5. TF-IDF provides better feature weighting.
6. Cosine similarity effectively identifies similar reviews.


### 10. Conclusion

In this experiment, One-Hot Encoding, Bag of Words, and TF-IDF were applied to a real-world Amazon product reviews dataset.
Preprocessing improved the quality of text representation.

TF-IDF proved to be the most effective technique, as it considers both term frequency and term importance.
Cosine similarity combined with TF-IDF produced meaningful document similarity results, making TF-IDF suitable for real-world NLP applications.
