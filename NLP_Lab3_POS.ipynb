### Part 1: Naive Bayes Classification

```python
import numpy as np
import pandas as pd
import time

from datasets import load_dataset
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

Load AG News Dataset

```python
dataset = load_dataset("ag_news")

train_texts = dataset["train"]["text"]
train_labels = dataset["train"]["label"]

test_texts = dataset["test"]["text"]
test_labels = dataset["test"]["label"]
```

Binary Bag of Words

```python
vectorizer_bin = CountVectorizer(binary=True)

X_train_bin = vectorizer_bin.fit_transform(train_texts)
X_test_bin = vectorizer_bin.transform(test_texts)

model_bin = BernoulliNB()
start = time.time()
model_bin.fit(X_train_bin, train_labels)
training_time_bin = time.time() - start

pred_bin = model_bin.predict(X_test_bin)
accuracy_bin = accuracy_score(test_labels, pred_bin)
f1_bin = f1_score(test_labels, pred_bin, average="macro")
```

Term Frequency

```python
vectorizer_tf = CountVectorizer(stop_words="english")

X_train_tf = vectorizer_tf.fit_transform(train_texts)
X_test_tf = vectorizer_tf.transform(test_texts)

model_tf = MultinomialNB()

start = time.time()
model_tf.fit(X_train_tf, train_labels)
training_time_tf = time.time() - start

pred_tf = model_tf.predict(X_test_tf)

acc_tf = accuracy_score(test_labels, pred_tf)
f1_tf = f1_score(test_labels, pred_tf, average="macro")
```

TF-IDF

```python
vectorizer_tfidf = TfidfVectorizer(stop_words="english")

X_train_tfidf = vectorizer_tfidf.fit_transform(train_texts)
X_test_tfidf = vectorizer_tfidf.transform(test_texts)

model_tfidf = MultinomialNB()

start = time.time()
model_tfidf.fit(X_train_tfidf, train_labels)
training_time_tfidf = time.time() - start

pred_tfidf = model_tfidf.predict(X_test_tfidf)

acc_tfidf = accuracy_score(test_labels, pred_tfidf)
f1_tfidf = f1_score(test_labels, pred_tfidf, average="macro")
```

Result Table

```python
results = pd.DataFrame({
    "Representation": ["Binary BoW", "TF", "TF-IDF"],
    "Accuracy": [accuracy_bin, acc_tf, acc_tfidf],
    "F1": [f1_bin, f1_tf, f1_tfidf],
    "Training Time": [training_time_bin, training_time_tf, training_time_tfidf]
})

print(results)
```

```python
import numpy as np
import pandas as pd

from datasets import load_dataset
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation


import gensim
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary
```

Load AG News

```python
dataset = load_dataset("ag_news")

texts = dataset['train']['text']
```

Preprocessing

```python
vectorizer = CountVectorizer(
    stop_words="english",
    max_df=0.95,
    min_df=5
)

X = vectorizer.fit_transform(texts)

features_names = vectorizer.get_feature_names_out()
```

Function to Train LDA

```python
def train_lda(n_topics):
  lda = LatentDirichletAllocation(
      n_components=n_topics,
      random_state=42,
      learning_method="batch"
  )

  lda.fit(X)
  perplexity = lda.perplexity(X)

  return lda, perplexity
```

Extract Top Words Per Topic

```python
def get_top_words(model, n_top_words=10):
  topics = []
  for topic_idx, topic in enumerate(model.components_):
    top_words = [features_names[i]
                 for i in topic.argsort()[:-n_top_words -1:-1]]
    topics.append(top_words)
    return topics
```

```python
tokenized_texts = [text.lower().split() for text in texts]

dictionary = Dictionary(tokenized_texts)

corpus = [dictionary.doc2bow(text) for text in tokenized_texts]
```

Compute Coherence

```python
def compute_coherence(topics):
  coherence_model = CoherenceModel(
      topics=topics,
      texts=tokenized_texts,
      dictionary=dictionary,
      coherence='c_v'
  )
  return coherence_model.get_coherence()
```

Train for 5, 10, 20 Topics

```python
results = []

for k in [5, 10, 20]:
  lda_model, perplexity = train_lda(k)
  topics = get_top_words(lda_model)
  coherence = compute_coherence(topics)

  results.append([k, perplexity, coherence])
```

Report Table

```python
results_df = pd.DataFrame(
    results,
    columns=["Topics", "Perplexity", "Coherence"]
)

results_df
```

```python
import nltk
import numpy as np
import pandas as pd
from collections import defaultdict, Counter
import math

nltk.download('treebank')
nltk.download('universal_tagset')
```

Load Penn Treebank

```python
from nltk.corpus import treebank

tagged_sentences = treebank.tagged_sents(tagset='universal')
```

```python
train_data = tagged_sentences[:3000]
test_data = tagged_sentences[3000:]
```

Build Baseline Model

```python
word_tag_freq = defaultdict(Counter)

for sentence in train_data:
    for word, tag in sentence:
        word_tag_freq[word][tag] += 1

# Build most frequent tag dictionary
most_freq_tag = {}

for word in word_tag_freq:
    most_freq_tag[word] = word_tag_freq[word].most_common(1)[0][0]
```

Baseline Prediction Function

```python
def baseline_predict(sentence):
    prediction = []
    for word in sentence:
        if word in most_freq_tag:
            prediction.append(most_freq_tag[word])
        else:
            prediction.append("NOUN")  # default for unseen
    return prediction
```

Compute Baseline Accuracy

```python
correct = 0
total = 0

for sentence in test_data:
    words = [word for word, tag in sentence]
    true_tags = [tag for word, tag in sentence]

    pred_tags = baseline_predict(words)

    for t, p in zip(true_tags, pred_tags):
        if t == p:
            correct += 1
        total += 1

baseline_accuracy = correct / total
print("Baseline Accuracy:", baseline_accuracy)
```

Estimate HMM Parameters

```python
transition_counts = defaultdict(Counter)
emission_counts = defaultdict(Counter)
tag_counts = Counter()

for sentence in train_data:
    prev_tag = "<START>"

    for word, tag in sentence:
        transition_counts[prev_tag][tag] += 1
        emission_counts[tag][word] += 1
        tag_counts[tag] += 1
        prev_tag = tag
```

```python
tags = list(tag_counts.keys())

def transition_prob(prev_tag, curr_tag):
    return (transition_counts[prev_tag][curr_tag] + 1) / \
           (sum(transition_counts[prev_tag].values()) + len(tags))

def emission_prob(tag, word):
    return (emission_counts[tag][word] + 1) / \
           (tag_counts[tag] + len(emission_counts[tag]))
```

Viterbi Algorithm

```python
def viterbi(sentence):
    V = [{}]
    path = {}

    # Initialization
    for tag in tags:
        V[0][tag] = math.log(transition_prob("<START>", tag)) + \
                    math.log(emission_prob(tag, sentence[0]))
        path[tag] = [tag]

    # Recursion
    for t in range(1, len(sentence)):
        V.append({})
        new_path = {}

        for curr_tag in tags:
            prob_tag = []

            for prev_tag in tags:
                prob = V[t-1][prev_tag] + \
                       math.log(transition_prob(prev_tag, curr_tag)) + \
                       math.log(emission_prob(curr_tag, sentence[t]))
                prob_tag.append((prob, prev_tag))

            best_prob, best_prev_tag = max(prob_tag)
            V[t][curr_tag] = best_prob
            new_path[curr_tag] = path[best_prev_tag] + [curr_tag]

        path = new_path

    # Termination
    final_prob, final_tag = max((V[len(sentence)-1][tag], tag) for tag in tags)

    return path[final_tag]
```

Evaluate HMM Accuracy

```python
correct = 0
total = 0

for sentence in test_data:
    words = [word for word, tag in sentence]
    true_tags = [tag for word, tag in sentence]

    pred_tags = viterbi(words)

    for t, p in zip(true_tags, pred_tags):
        if t == p:
            correct += 1
        total += 1

hmm_accuracy = correct / total
print("HMM Accuracy:", hmm_accuracy)
```

```python
results = pd.DataFrame({
    "Model": ["Baseline", "HMM"],
    "Tagging Accuracy": [baseline_accuracy, hmm_accuracy]
})

results
```
